**Monte Carlo: Acierto y Error**

Gómez Cruz Manuel

Es una forma eficaz para generar muestras de una variable de distribución padre f(x) y contiene el germen de las ideas para técnicas más elaboradas, así como para introducir el

uso de las cadenas de Markov.

**Pasos:**
1. Generar Y parecido a g(.)
2. Generar un número aleatorio U uniforme en 0,1.
3. Si U<f(Y)/cg(Y), X=Y. Otro caso ir al paso 1.

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd

#Definir la función
def f(x):
  return np.sqrt(np.arctan(x))
#Graficamos función
x=np.linspace(0,1,400)
y=f(x)

#Definimo la función de los puntos aleatorios
#Ya no puse los puntos porque tardaba más en correr
def dardos (n):
  exitos=0
  for i in range(n):
    u1=rd.random()
    u2=rd.random()
    if u2<=f(u1):
      exitos+=1
  prob=exitos/n
  return prob
Graficamos las rectas donde queremos la integral
plt.plot(x,y)
plt.axhline(f(1),color="red")
plt.axvline(1,color="red")

I=dardos(10000)
print("La aproximación de la integral I es: ",I)
```

```python
def experimento(m):
  lista=[]
  for i in range (m):
    d=dardos(i+1)
    lista.append(d)
  return lista

L=experimento(300)
plt.plot(L)
print("La media es: ", np.mean(L))
print("La varianza es: ",np.var(L))
plt.axhline(np.mean(L),color="red")
plt.axhline(np.mean(L)+np.std(L),color="green")
plt.axhline(np.mean(L)-np.std(L),color="green")
plt.legend(["Media","Varianza","Intervalo de confianza"])
plt.show()
```

```python
from sympy import integrate, Symbol, sqrt, atan

x = Symbol('x')

#Integral exacta
resultado = integrate(sqrt(atan(x)), (x, 0, 1))

#Evaluando en resultado
exacta = resultado.evalf()

print("Solución exacta: ",exacta)

#Error
error=abs(exacta-I)
print("El error de la integral es: ",error)
```
