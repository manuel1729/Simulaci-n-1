# Funciones de distribución Continuas

### 1. Uniforme Continua $U(a, b)$

- **Nomenclatura**: U(a, b)
- **Función de densidad (PDF)**:
  \[
  f(x) = \begin{cases}
    \frac{1}{b-a} & \text{para } a \leq x \leq b \\
    0 & \text{en otro caso}
  \end{cases}
  \]

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/9/96/Uniform_Distribution_PDF_SVG.svg/500px-Uniform_Distribution_PDF_SVG.svg.png" width = "250"/>
</div>

- **Función acumulativa (CDF)**:
  $
  F(x) = \begin{cases}
    0 & \text{para } x < a \\
    \frac{x-a}{b-a} & \text{para } a \leq x \leq b \\
    1 & \text{para } x > b
  \end{cases}
  $

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/6/63/Uniform_cdf.svg/500px-Uniform_cdf.svg.png" width = "250"/>
</div>

- **Media**:
  $
  \mu = \frac{a + b}{2}
  $
- **Varianza**:
  $
  \sigma^2 = \frac{(b - a)^2}{12}
  $
- **Desviación estándar**:
  $
  \sigma = \frac{b - a}{\sqrt{12}}
  $
---

```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
import random as rd
#Función que repite n veces valor de x
def uniforme(a,b, n):
  x = []
  for i in range(n):
    U = rd.random()
    x.append(a + (b-a)*U)
  return x
#Graficar histograma,sacamos media y varianza
exp = uniforme(0,1,10000)
plt.hist(exp, bins=100)
print("Media aprox: ",np.mean(exp))
print("Varianza aprox: ",np.var(exp))
```

## 2. Triangular $T(a,b,c)$


  NOMENCLATURA: T(a,c,b)  
- **PDF (Función de Densidad):**

$f(x)=\left\{\begin{array}{lc} \frac{2}{(c-a)(b-a)}(x-a) & a\leq x \leq b\\ \frac{2}{(c-a)(b-c)}(x-c) & b\leq x\leq c \end{array}\right.$

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/4/45/Triangular_distribution_PMF.png/330px-Triangular_distribution_PMF.png" width = "250"/>
</div>

- **CDF (Función de Distribución Acumulativa):**  
$F(x)= \left\{\begin{array}{lc}0 & \text{para } x < a \\ \frac{(x-a)^2}{(c-a)(b-a)} & a\leq x \leq c \\ 1+\frac{(x-c)^2}{(c-a)(b-c)} & c\leq x\leq b \\
1 & \text{para } x > b \end{array}\right.$

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/e/e5/Triangular_distribution_CMF.png/330px-Triangular_distribution_CMF.png" width = "250"/>
</div>

$\mu=\frac{a+b+c}{3} \qquad \sigma^2=\frac{a^2+b^2+c^2-ab-ac-bc}{18}$

**PSEUDOCÓDIGO**

1. Generate U~U(0, 1).
2. if $U\leq M=\frac{b-a}{c-a}$

      return $X=\sqrt{U(c-a)(b-a)}+a$

   else:

      return $X=c- \sqrt{(U-1)(c-a)(b-c)}$
---

```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función que repite n veces valor de x
def triangular(a,b,c,n):
  x = []
  for i in range(n):
    U = random()
    if U <= (b-a)/(c-a):
      x.append(np.sqrt(U*(c-a)*(b-a))+a)
    else:
      x. append(-np.sqrt((U-1)*(c-a)*(b-c))+c)
  return x

#Graficar histograma,sacamos media y varianza
exp = triangular(1,3,5,100000)
plt.hist(exp, bins=100)
print("Media aprox: ",np.mean(exp))
print("Varianza aprox: ",np.var(exp))
```

## 3. Exponencial (Exp($λ=\frac{1}{β}$))

- **Nomenclatura**: Exp($λ=\frac{1}{β}$)
- **Función de densidad (PDF)**:

  $f(x) = \begin{cases}
    \lambda e^{-\lambda x} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
  \end{cases}$

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/e/ec/Exponential_pdf.svg/800px-Exponential_pdf.svg.png" width = "250"/>
</div>

- **Función acumulativa (CDF)**:
  
  $F(x) = \begin{cases}
    1 - e^{-\lambda x} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
  \end{cases}
  $

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/b/ba/Exponential_cdf.svg/800px-Exponential_cdf.svg.png" width = "250"/>
</div>

- **Media**:

  $\mu = \frac{1}{\lambda}$
  
- **Varianza**:
  
  $\sigma^2 = \frac{1}{\lambda^2}$
  
- **Desviación estándar**:
  
  $\sigma = \frac{1}{\lambda}$
---

  **PSEUDOCÓDIGO**

1. Generate U~U(0, 1).
2. Return $X=-\mu \cdot ln(U)=-\frac{ln(U)}{\lambda}$
---

```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función que repite n veces valor de x
def exponencial(lanbda, n):
  x = []
  for i in range(n):
    R = random()
    x. append(-np.log(R)/lanbda)
  return x

#Graficar histograma,sacamos media y varianza
exp = exponencial(1.5,10000)
plt.hist(exp, bins=100)
print("Media aprox:",np.mean(exp))
print("Varianza aprox: ",np.var(exp))
```

## 4. m-Erlang (Erlang(m, λ))

- **Nomenclatura**: Erlang(m, λ)
- **Función de densidad (PDF)**:
  
  $
  f(x) = \begin{cases}
    \frac{\lambda^m x^{m-1} e^{-\lambda x}}{(m-1)!} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}
  $
  <div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/f/fc/Gamma_distribution_pdf.png/800px-Gamma_distribution_pdf.png" width = "250"/>
</div>
- **Función acumulativa (CDF)**:

  $
  F(x) = \begin{cases}1 - e^{-\lambda x}\sum_{k=0}^{m-1} \frac{(\lambda x)^k}{k!}& \text{para } x > 0\\ 0 & \text{otro caso} \end{cases}
  $
  <div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/a/a9/Gamma_distribution_cdf.png/800px-Gamma_distribution_cdf.png" width = "250"/>
</div>
- **Media**:
  $
  \mu = \frac{m}{\lambda}
  $
- **Varianza**:
  $
  \sigma^2 = \frac{m}{\lambda^2}
  $
- **Desviación estándar**:
  $
  \sigma = \frac{\sqrt{m}}{\lambda}
  $
---
  **PSEUDOCÓDIGO**
  
1- Generar $U_1,U_2,...,U_m$ como IID (Independiente e Independientemente distribuido) U(0,1)

2-Return $X=-\frac{\mu}{m}ln(\prod_{i=1}^{m}U_i)=\frac{ln(\prod_{i=1}^{m}U_i}{\lambda}$

```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función que repite n veces valor de x
def erlang(m, lanbda,n):
  x = []
  mu = m / lanbda #Media
  for i in range(n):
    U = [random() for _ in range(m)]
    producto = np.prod(U)
    x.append(-(mu / m) * np.log(producto))
  return x

#Graficar histograma,sacamos media y varianza
exp = erlang(10, 1.5,10000)
plt.hist(exp, bins=100)
plt.show()
print("Media aprox:", np.mean(exp))
print("Varianza aprox: ", np.var(exp))
```

## 5. Gamma (α, β)

- **Nomenclatura**: Gamma(α, β)

Tomando $m=\alpha \qquad λ=\frac{1}{β}$

$\frac{\lambda^m x^{m-1} e^{-\lambda x}}{(m-1)!} ⇒ \frac{\lambda^m x^{m-1} e^{-\lambda x}}{\Gamma(m)}$
- **Función de densidad (PDF)**:
  
  $
  f(x) = \begin{cases}
    \frac{\beta^{-\alpha} x^{\alpha-1} e^{-\frac{x}{\beta}}}{\Gamma(\alpha)} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}
  $
  <div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/f/fc/Gamma_distribution_pdf.png/800px-Gamma_distribution_pdf.png" width = "250"/>
</div>
- **Función acumulativa (CDF)**:

 $
  F(x) = \begin{cases}1 - e^{-\frac{x}{\beta}}\sum_{j=0}^{
  \alpha-1} \frac{(\frac{x}{β})^j}{j!}& \text{para } x > 0\\ 0 & \text{otro caso} \end{cases}
  $
  <div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/a/a9/Gamma_distribution_cdf.png/800px-Gamma_distribution_cdf.png" width = "250"/>
</div>
- **Media**:
  
  $
  \mu = {\alpha}{\beta}
  $
- **Varianza**:
  $
  \sigma^2 = {\alpha}{\beta^2}
  $
- **Desviación estándar**:
  $
  \sigma = {\sqrt{\alpha}}\cdot{\beta}
  $
---


**PSEUDOCÓDIGO**

**Para $0<\alpha< 1$**

con $b=\frac{e+\alpha}{e}$

1. Generar $U_1$~$U(0,1)$ y sea $P=bU_1$
  
  if P>1, ir al paso 3
  
  otro caso: ir al paso 2

2. Sea $Y=P^{\frac{1}{\alpha}}$ y generar $U_2$~$U(0,1)$
   
   if $U_2\leq e^{-Y}$

      return $X=Y$
   otro caso: ir al paso 1.

3. Sea $Y=-ln(\frac{b-P}{\alpha})$ y generar $U_2$~$U(0,1)$

    if $U_2\leq Y^{\alpha-1}$

      return $X=Y$

    otro caso: ir al paso 1.
---

```python
#GAMMA CON 0<alpha<1 y beta=1
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función que repite n veces valor de x
def gamma_1(alpha,n):
    if not (0 < alpha < 1):
        raise ValueError("Válido para 0 < alpha < 1.")

    e = np.e
    b = (e + alpha) / e
    x=[]
    for i in range(n):
      U1 = random()
      P = b * U1

      if P <= 1:
        Y = P ** (1 / alpha)
        U2 = random()
        if U2 <= np.exp(-Y):
          x.append(Y)
      else:
        Y = -np.log((b - P) / alpha)
        U2 = random()
        if U2 <= Y ** (alpha - 1):
          x.append(Y)
    return x

#Graficar histograma,sacamos media y varianza
exp = gamma_1(0.5,10000)
plt.hist(exp, bins=100)
print("Media aprox:",np.mean(exp))
print("Varianza aprox: ",np.var(exp))
```

PARA $\alpha>1$

$a=\frac{1}{\sqrt{2\alpha -1}}$

$b=\alpha-ln(4)$

$q=\alpha+\frac{1}{a}$

$\theta =4.5$

$d=1+ln(θ)$

1. Generar $U_1,U_2$ como IID $U(0,1)$

2. Sea $V=a\cdot ln(\frac{U_1}{1-U_1}) \qquad Y=\alpha e^V \qquad Z=U_1^2U_2 \qquad W=b+qV-Y$

3. if $W+d-θZ\geq 0$

     return $X=Y$

    otro caso: ir al paso 4

4. if $W\geq ln(Z)$
    
    return $X=Y$

  otro caso: ir al paso 1

```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función que repite n veces valor de x
def gamma_2(alpha,n):
  if (0 < alpha < 1):
    raise ValueError("Válido para alpha > 1.")

  a = 1 / np.sqrt(2 * alpha - 1)
  beta = alpha - np.log(4)
  q = alpha + 1 / a
  theta = 4.5
  d = 1 + np.log(theta)

  x = []
  for i in range(n):
    U1 = random()
    U2 = random()
    V = a * np.log(U1 / (1 - U1))
    Y = alpha * np.exp(V)
    Z = U1 ** 2 * U2
    W = beta + q * V - Y
    if W + d - theta * Z >= 0:
      x.append(Y)
    elif W >= np.log(Z):
      x.append(Y)
  return x

#Graficar histograma,sacamos media y varianza
exp = gamma_2(2,10000)
plt.hist(exp, bins=100)
print("Media aprox:",np.mean(exp))
print("Varianza aprox: ",np.var(exp))
```

## 6. Weibull($\alpha$,$β=\frac{1}{\lambda}$)

- **Nomenclatura**: Weibull($\alpha$,$β=\frac{1}{\lambda}$)
- **Función de densidad (PDF)**:
  
  $
  f(x) = \begin{cases}
    αβ^{-α}\cdot{x}^{α-1} e^{-(\frac{x}{β})^α} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}
  $
  <div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/5/58/Weibull_PDF.svg/800px-Weibull_PDF.svg.png" width = "250"/>
</div>
- **Función acumulativa (CDF)**:
  
  $
  F(x) = \begin{cases}
    1 - e^{-(\frac{x}{β})^α} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}
  $
  <div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/7/7e/Weibull_CDF.svg/800px-Weibull_CDF.svg.png" width = "250"/>
</div>
- **Media**:
  
  $
  \mu = \frac{β}{α} \cdot \Gamma\left(\frac{1}{α}\right)
  $
- **Varianza**:
  
  $
  \sigma^2 = \frac{\beta^2}{\alpha} \left[ 2\Gamma\left(\frac{2}{α}\right) -\frac{1}{α} \left( \Gamma\left(\frac{1}{α}\right) \right)^2 \right]
  $
- **Desviación estándar**:
  
  $
  \sigma = \sqrt{\sigma^2}
  $
---
**PSEUDOCÓDIGO**
1. Generar $U$~$U(0,1)$
2. return $X=β(-ln(U))^{\frac{1}{\alpha}}$

```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función que repite n veces valor de x
def weibull(beta,alpha, n):
  x = []
  for i in range(n):
    U = random()
    x. append(beta*(-np.log(U))**(1/alpha))
  return x

#Graficar histograma,sacamos media y varianza
exp = weibull(1,3,10000)
plt.hist(exp, bins=100)
print("Media aprox:",np.mean(exp))
print("Varianza aprox: ",np.var(exp))
```

## 7. Normal (N(μ, σ²))

- **Nomenclatura**: N(μ, σ²)
- **Función de densidad (PDF)**:
  
  $
  f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
  \quad ∀ x \in \mathcal{R}$
<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/1/1b/Normal_distribution_pdf.png/800px-Normal_distribution_pdf.png" width = "250"/>
</div>

- **Función acumulativa (CDF)**:
  $
  F(x) = \Phi\left( \frac{x - \mu}{\sigma} \right)
  $
<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/1/19/Normal_distribution_cdf.png/800px-Normal_distribution_cdf.png" width = "250"/>
</div>
- **Media**:
  $
  \mu = \mu
  $
- **Varianza**:
  $
  \sigma^2 = \sigma^2
  $
- **Desviación estándar**:
  $
  \sigma = \sigma
  $
---

**PSEUDOCODIGO**

1. Generar $U_1,U_2\sim (0,1)$

2. Hacer $R=\sqrt{-2ln(U_1)}$
      
      $\qquad θ=2\pi U_2$

      Elegir cualquiera
3. return

      $\quad X=Rcos(θ)$  
      $\quad Y=Rsen(θ)$

---

```python
#Librerías
import random as rd
import numpy as np
import matplotlib.pyplot as plt

#Experimento para repetir la prueba n veces
#y agregar a lista
def iteraciones(n):
  lista=[]
  for i in range(n):
    U1=rd.random()
    U2=rd.random()
    R=np.sqrt(-2*np.log(U1))
    theta=2*np.pi*U2
    x=R*np.cos(theta)
    y=R*np.sin(theta)
    lista.append(x)
  return lista

#Para n=10000
l=iteraciones(10000)
plt.hist(l, bins=100)
print("Media aprox:",np.mean(l))
print("Varianza aprox: ",np.var(l))
```

## 8. Log-normal(LN(μ, σ²))

- **Nomenclatura**: LN(μ, σ²)
- **Función de densidad (PDF)**:
  $
  f(x) = \begin{cases}
    \frac{1}{x \sigma \sqrt{2\pi}} e^{-\frac{(\ln x - \mu)^2}{2\sigma^2}} & \text{para } x > 0 \\
    0 & \text{para } x \leq 0
  \end{cases}
  $
  <div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/8/89/Log-normal-pdfs.png/330px-Log-normal-pdfs.png" width = "250"/>
</div>
- **Función acumulativa (CDF)**:
  $
  F(x) = \Phi\left( \frac{\ln x - \mu}{\sigma} \right)$
<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/dc/Log-normal-cdfs.png/330px-Log-normal-cdfs.png" width = "250"/>
</div>
- **Media**:
  $\mu = e^{\mu + \frac{\sigma^2}{2}}$

- **Varianza**:
  $
  \sigma^2 = (e^{\sigma^2} - 1) e^{2\mu + \sigma^2}
  $

- **Desviación estándar**:
  $
  \sigma = \sqrt{(e^{\sigma^2} - 1) e^{2\mu + \sigma^2}}
  $
---
**PSEUDOCÓDIGO**

Si $X\sim LN(μ, σ²)$ y sea $\mu´=E(X)$ y $\sigma´^2=Var(X)$

$⇒ \mu´= e^{\mu + \frac{\sigma^2}{2}} \qquad \sigma´^2=(e^{\sigma^2} - 1) e^{2\mu + \sigma^2}$

Dados $\mu´$ y $\sigma´^2$ media y varianza deseada a simular se calcula

$\mu=ln( \frac{\mu´^2}{\sqrt{\mu´^2+\sigma´^2}})$

$\sigma^2=ln(1+\frac{\sigma´^2}{\mu´^2})$
1. Generar $Y\sim N(μ, σ²)$
2. return $X=e^Y$
---

```python
#Librerías
import random as rd
import numpy as np
import matplotlib.pyplot as plt

#Experimento para repetir la prueba n veces
#y agregar a lista
def log_normal(muprima,sigmaprima,n):
  mu=np.log(muprima**2/np.sqrt(muprima**2+sigmaprima**2))
  sigma=np.sqrt(np.log(1+sigmaprima**2/muprima**2))
  X=[]
  for i in range(n):
    Y=np.random.normal(mu,sigma)
    X.append(np.exp(Y))
  return X

#Graficar histograma,sacamos media y varianza
l=log_normal(1,1,1000)
plt.hist(l, bins=100)
print("Media aprox:",np.mean(l))
print("Varianza aprox: ",np.var(l))
```

## 9. Beta(($α_1, α_2$))

- **Nomenclatura**: Beta($α_1, α_2$)
- **Función de densidad (PDF)**:
  $
  f(x) = \begin{cases}
    \frac{x^{\alpha_1-1} (1-x)^{\alpha_2-1}}{B(\alpha_1, \alpha_2)} & \text{para } 0 \leq x \leq 1 \\
    0 & \text{en otro caso}
  \end{cases}
  $
<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/9/9a/Beta_distribution_pdf.png/330px-Beta_distribution_pdf.png" width = "250"/>
</div>
  Donde $B(\alpha_1, \alpha_2)$ es la *beta funcion*, definida como
  
  $B(z_1,z_2)=\int_{0}^{1}t^{z_1-1}(1-t)^{z_2-1}\cdot dt$

  para números reales $z_1>0$ y $z_2>0$

- **Función acumulativa (CDF)**: No tiene forma cerrada general.
- **Media**:
  $\mu = \frac{\alpha}{\alpha + \beta}$
<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/7/76/Beta_distribution_cdf.png/330px-Beta_distribution_cdf.png" width = "250"/>
</div>

- **Varianza**:
  $\sigma^2 = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}$

- **Desviación estándar**:

  $\sigma = \sqrt{\frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}}$
---
**PSEUDOCÓDIGO**

1. Generar $Y_1\sim$gamma($\alpha_1$,1) y $Y_2\sim$gamma($\alpha_2$,1) independiente de $Y_1$
2. Regresar $X=\frac{Y_1}{Y_1+Y_2}$

```python
#Librerías
import random as rd
import numpy as np
import matplotlib.pyplot as plt


#Experimento para repetir la prueba n veces
#y agregar a lista
def beta(alpha1,alpha2,n):
  X=[]
  for i in range(n):
    Y1=np.random.gamma(alpha1,1)
    Y2=np.random.gamma(alpha2,1)
    X.append(Y1/(Y1+Y2))
  return X

#Graficar histograma,sacamos media y varianza
k=beta(0.5,0.5,1000)
plt.hist(k, bins=50)
print("Media aprox:",np.mean(k))
print("Varianza aprox: ",np.var(k))
```

##10. Pearson Tipos

**A) Pearson Tipo V**

- **Nomenclatura**: PT5(α, β)
- **Función de densidad (PDF)**:
  \[
  f(x) = \begin{cases}
    \frac{x^{-(\alpha-1)}e^{-\frac{\beta}{x}}}{β^{-\alpha}ℾ(α)} & \text{para } x > 0 \\
    0 & \text{otra caso}
  \end{cases}
  \]
  1000007934.png
- **Función acumulativa (CDF)**:
  \[
  F(x) = \begin{cases} 1-
   F_G (\frac{1}{x}) & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}
  \]
- **Media**:
  $
  \mu = \frac{β}{\alpha -1}
  $  para $\alpha>1$

- **Varianza**:
  $
  \sigma^2 = \frac{\beta^2}{(\alpha-1)^2(α-2)}
  $   para $\alpha>2$
- **Desviación estándar**:
  $
  \sigma = \sqrt{\frac{\beta^2}{(\alpha-1)^2(α-2)}}
  $
---

**PSEUDOCÓDIGO**

$X\sim PT5(\alpha,β)$ si y solo si $\frac{1}{X}\sim gamma(\alpha,\frac{1}{β})$

1. Generar $Y∼gamma(\alpha,\frac{1}{β})$

2. Return $X=\frac{1}{Y}$

```python
#Librerias
from random import random
import numpy as np
import matplotlib.pyplot as plt

#Experimento para repetir la prueba n veces
#y agregar a lista
def pearson_v(alpha,beta,n):
  x=[]
  for i in range(n):
    Y=np.random.gamma(alpha,1/beta)
    x.append(1/Y)
  return x

#Graficar histograma,sacamos media y varianza
exp = pearson_v(2,1,10000)
plt.hist(exp, bins=100,range=(0,5))
print("Media aprox:",np.mean(exp))
print("Varianza aprox: ",np.var(exp))
```

**B) Pearson Tipo VI**

- **Nomenclatura**: PT6($α_1,α_2 β$)
- **Función de densidad (PDF)**:

2. Generar $U_{i+1}\sim U(0,1)$
3. remplazar  $bU_{i+1}$
4. Si $b<a$

   return $X=i$

```python
#librerias
import random as rd
import numpy as np
import matplotlib.pyplot as plt

#Experimento para repetir la prueba n veces
#y agregar a lista
def poisson(lam,n):
  l=[]
  for i in range(n):
    a=np.exp(-lam)
    b=1
    i=0
    while True:
        u = np.random.uniform(0, 1)
        b *= u
        if b < a:
          break
        i += 1
    l.append(i)
  return l

#Graficar histograma,sacamos media y varianza
exp=poisson(2,1000)
plt.hist(exp, bins=100)
print("Media aprox:",np.mean(exp))
print("Varianza aprox: ",np.var(exp))
```

#Teoremas proba

## Límite Central

Sea \( $X_1, X_2, \dots, X_n $\) una sucesión de variables aleatorias independientes e idénticamente distribuidas (i.i.d.) con media \($ \mu $\) y varianza \($ \sigma^2 $\), entonces la suma estandarizada:

$$
Z_n = \frac{\bar{X}_n - \mu}{\sigma / \sqrt{n}}
$$

donde \( $\bar{X}n = \frac{1}{n} \sum{i=1}^{n} X_i $\), converge en distribución a una normal estándar:

$$
Z_n \xrightarrow{d} \mathcal{N}(0, 1) \quad \text{cuando } n \to \infty
$$

Esto implica que, para valores suficientemente grandes de \( n \), la distribución de \( $\bar{X}_n$ \) se aproxima a una distribución normal:

$$
\bar{X}_n \sim \mathcal{N}\left(\mu,\ \frac{\sigma^2}{n} \right)
$$

Válido incluso si la distribución original de las ($ X_i $\) no es normal, siempre que se tengan media y varianza finitas.

## Resta de dos variables aleatorias

Si X y Y son variables aleatorias y independientes, entonces definimos:

* La esperanza de la resta:

  $$
  E[X - Y] =E[X] -E[Y]
  $$

* Varianza
  $$
  Var(X - Y) = \sigma^2_X + \sigma^2_Y
  $$
  
*  Si $X \sim N$($\mu_X, \sigma_X^2$) \) y \($ Y \sim N$($\mu_Y, \sigma_Y^2$) \), entonces:

  $$
  X - Y \sim N(\mu_X - \mu_Y,\ \sigma_X^2 + \sigma_Y^2)
  $$
