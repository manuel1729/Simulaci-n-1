**MONTE CARLO CRUDO (Media Muestral)**
I=∫g(x)dx=∫w(x)⋅f(x)dx=E[g(x)]≈(1/N)∑_{i=1}_{N}g(x_i)

donde w(x)=g(x)(b-a) y f(x)=1/(b-a) que es la distriobución uniforme U(a,b)

**Pasos:**

1. Generar secuencia {U_i}_{i=1}_{N}
2. Calcular X_i=a+U_i(b-a)
3. Calcular g(X_i)
4. Estimar I con la sumatoria

```python
import numpy as np
import math
import random
import matplotlib.pyplot as plot
#Funcion g(x)
def g(x):
  return math.sqrt(math.atan(x))

def monte_carlo_crudo(g,a,b,n):
  al=[]
  x=[]
  for i in range (n):
    #Paso 1
    al.append(random.uniform(a,b))
    #Paso 2 y 3
    x.append(g((al[i]*(b-a))+a))
    #Paso 4
  return (sum(x)/n)*(b-a)

I=monte_carlo_crudo(g,0,1,1000)
print("El resuldo de I: ",I)
```
