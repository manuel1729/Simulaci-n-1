El generador Congruencial Lineal de Lehmer

Propuesto por D. H. Lehmer, en el cual se escogen 4 números mágicos del conjunto de los números enteros no negativos así:

X_0>=0
Con X0 cualquier entero positivo menor a "m"

a, el multiplicador a >= 0
Con (a-1) múltiplo de p para cada primo p que divide a "m"

c,  el incremento c >=0
con c un entero positivo menor que m y primo relativo con m (función de euler)

m,  el módulo m > X_0 , m > a, m > c  

con m cualquier entero positivo (un número grande), ya que coincide con el máximo periodo

La secuencia deseada de números aleatorios es obtenida colocando  

X_{j+1} = (aX_j + c) mod m

Esta es conocida como una secuencia lineal de congruencias o generador lineal de congruencias,
y se denota con GLC (\( X_0, a, c, m \)

```python
#Gomez Cruz Manuel
import tabulate
import matplotlib.pyplot as plt

# Parámetros del generador lineal congruencial
a = 81
c = 89
m = 100
x0 = 5

# Función generadora
def generador(a, c, m, x0):
    x = (a * x0 + c) % m
    return x

# Listas para almacenar los valores generados
muestra = []
lista = []
valores_vistos = set()  # Conjunto para almacenar valores ya vistos
ciclo_detectado = False
ciclo_vida = 0

# Generamos la tabla de valores
for i in range(m+1):
    nuevo = generador(a, c, m, x0)
    muestra.append([i, x0, nuevo / m])
    lista.append([nuevo / m])

    # Verificamos si el valor ya ha sido visto
    if nuevo in valores_vistos:
        ciclo_detectado = True
        ciclo_vida = i
        break
    else:
        valores_vistos.add(nuevo)

    x0 = nuevo

# Imprimimos la tabla con los valores generados
encabezados = ["i", "x_i", "x_[i+1]"]
print(tabulate.tabulate(muestra, headers=encabezados))


# Imprimir el ciclo de vida si se detectó
if ciclo_detectado:
    print(f"El ciclo de vida es: {ciclo_vida}")
else:
    print("No se detectó un ciclo dentro del rango de iteraciones.")
```
